```python
# import modules
import uproot, sys, time, random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import awkward as ak
from tqdm.rich import tqdm
import seaborn as sns
from matplotlib.ticker import FormatStrFormatter
import matplotlib.ticker as ticker
from collections import Counter
import plotly.graph_objects as go

# Set up plot defaults
import matplotlib as mpl
mpl.rcParams['figure.figsize'] = 14.0,10.0  # Roughly 11 cm wde by 8 cm high
mpl.rcParams['font.size'] = 20.0 # Use 14 point font
sns.set(style="whitegrid")

font_size = {
    "xlabel": 17,
    "ylabel": 17,
    "xticks": 15,
    "yticks": 15,
    "legend": 13,
    "title": 13,
}

plt.rcParams.update({
    "axes.labelsize": font_size["xlabel"],  # X and Y axis labels
    "xtick.labelsize": font_size["xticks"],  # X ticks
    "ytick.labelsize": font_size["yticks"],  # Y ticks
    "legend.fontsize": font_size["legend"]  # Legend
})

```

```python
def plot(path):
    path_athena = path + 'strip_cluster_athena.csv'
    path_traccc = path + 'strip_cluster_traccc.csv'
    dfa = pd.read_csv(path_athena)
    dfc = pd.read_csv(path_traccc)

    dfa['rank'] = dfa.groupby("athena_id").cumcount()
    dfc['rank'] = dfc.groupby("athena_id").cumcount()

    # Merge
    merged = pd.merge(dfa, dfc, on=["athena_id", "rank"], suffixes=("_athena", "_traccc"))

    merged["residual_local_x"] = merged["local_x_traccc"] - merged["local_x_athena"]
    merged["residual_local_y"] = merged["local_y_traccc"] - merged["local_y_athena"]

    groups = {
        "-2": merged[merged["barrel_ec_athena"] == -2],
        "0": merged[merged["barrel_ec_athena"] == 0],
        "2": merged[merged["barrel_ec_athena"] == 2],
    }

    for label, df in groups.items():
        plt.figure(figsize=(12, 5))

        plt.subplot(1, 2, 1)
        data = df["residual_local_x"]
        plt.hist(data, bins=100, alpha=0.7, label=f'μ:{np.mean(data):.2g}, σ:{np.std(data):.2g}')
        plt.xlabel("residual_local_x [mm]")
        plt.ylabel("count")
        plt.title(f"Residual Local X for barrel_ec = {label}")
        plt.legend()

        plt.subplot(1, 2, 2)
        data = df["residual_local_y"]
        plt.hist(data, bins=100, alpha=0.7, label=f'μ:{np.mean(data):.2g}, σ:{np.std(data):.2g}', color="orange")
        plt.xlabel("residual_local_y [mm]")
        plt.ylabel("count")
        plt.title(f"Residual Local Y for barrel_ec = {label}")
        plt.legend()

        plt.tight_layout()
```
